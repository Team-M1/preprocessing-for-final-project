# 1. 혐오표현.csv와 kocohub_labeled_data.csv 합치기

합치기의 기준은 혐오표현.csv  
kocohub데이터에서 contain_gender_bias == True이고 hate == "hate"인 데이터만 gender_hate를 1로 준다.

```python
import pandas as pd


data1 = pd.read_csv("./data/혐오표현.csv")
data2 = pd.read_csv("./data/kocohub_labeled_data.csv")
```

```python
data1.head()
```

```python
data2.head()
```

```python
# "contain_gender_bias"가 True이고 "hate"가 "hate"인 행의 인덱스

idx = (data2["contain_gender_bias"] == True) & (data2["hate"] == "hate")
```

```python
# gender_hate 열을 만들고 위 조건에 맞는 행을 1로, 나머지를 0으로 한다.

data2["gender_hate"] = 0
data2.loc[idx, "gender_hate"] = 1
data2.head()
```

```python
# 같은 방법으로 hate2 열을 만들고 "hate"가 "hate"인 행을 1로 한다.

data2["hate2"] = 0
data2.loc[data2["hate"] == "hate", "hate2"] = 1
data2.head()
```

```python
# 혐오표현.csv와 형식을 맞추기 위해 새로 만든 열과 댓글 내용만 뽑는다.

new_data2 = data2.loc[:, ["comments", "hate2", "gender_hate"]]
new_data2.head()
```

```python
# 열 이름을 똑같이 맞춘다.

new_data2.columns = ["content", "hate", "gender_hate"]
new_data2.head()
```

```python
# 합침

con_data = pd.concat([data1, new_data2])
con_data.shape
```

```python
from data_preprocess import text_preprocessing


con_data["content"] = con_data["content"].apply(text_preprocessing)
con_data.head()
```

```python
# 저장

con_data.to_csv("./data/혐오표현_정제.csv", index=False)
```
